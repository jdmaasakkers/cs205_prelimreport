```python
#Initialize arrays
size_array = np.array([1000,5000,10000,20000])
time_array_od = np.array([83.70519924163818,416.01120805740356,835.3090200424194,1669.194750547409])
time_array_lt = np.array([48.8767158985 ,233.836622,459.200274944])

#Plot
plt.plot(size_array,time_array_od, label=r'Timed serial code (Odyssey)',c='purple')
#plt.plot(size_array[:-1],time_array_lt, label=r'Timed serial code (Laptop)',c='pink')

plt.xlabel(r'Samples',fontsize=15)
plt.ylabel(r'Runtime',fontsize=15)
plt.title(r'Runtime as a function of number of samples',fontsize=16)
plt.tick_params(direction='out', length=5, width=1, which='major', colors='black', top='off')
#plt.yscale('log')
plt.xlim([0,size_array[-1]*1.1])
plt.xticks(size_array,fontsize=14)
plt.yticks(fontsize=14)
plt.legend(loc='upper left', fontsize = 15)
plt.show()
```

```python
time_array_1000 = np.array([1.2452025413513184,0.9349987506866455,0.7121853828430176,0.8527417182922363])
time_array_5000 = np.array([6.274763822555542,4.489902019500732,3.3720932006835938,2.998167037963867])
time_array_10000 = np.array([12.567538976669312,8.999614715576172,6.816726922988892,6.097769737243652])
time_array_20000 = np.array([25.052375078201294,17.94033932685852,13.880289316177368,12.212865829467773])
```

```python
#Initialize arrays
p_array = np.array([1,2,5,10])

plt.plot(p_array,time_array_1000, label=r'n = 1000',c='red')
plt.plot(p_array,time_array_5000, label=r'n = 5000',c='purple')
plt.plot(p_array,time_array_10000, label=r'n = 10000',c='blue')
plt.plot(p_array,time_array_20000, label=r'n = 20000',c='pink')

plt.xlabel(r'Cores',fontsize=15)
plt.ylabel(r'Time',fontsize=15)
plt.title(r'Runtime as a function of number of cores',fontsize=16)
plt.tick_params(direction='out', length=5, width=1, which='major', colors='black', top='off')
#plt.yscale('log')
plt.xlim([0,p_array[-1]*1.1])
plt.xticks(p_array,fontsize=14)
plt.yticks(fontsize=14)
plt.legend(loc='upper right', fontsize = 15)
plt.show()
```

```python
#Initialize arrays
p_array = np.array([1,5,10])

plt.plot(p_array,np.array([time_array_1000[0],time_array_1000[0],time_array_1000[0]])/np.array([time_array_1000[0],time_array_1000[2],time_array_1000[3]]), label=r'Speedup',c='silver')
plt.plot(p_array,np.array([time_array_1000[0],time_array_5000[0],time_array_10000[0]])/np.array([time_array_1000[0],time_array_5000[2],time_array_10000[3]]), label=r'Scaled speedup',c='gold')

plt.xlabel(r'Cores',fontsize=15)
plt.ylabel(r'Speedup',fontsize=15)
plt.title(r'Speedups as a function of number of cores (Basecase is n = 1000)',fontsize=16)
plt.tick_params(direction='out', length=5, width=1, which='major', colors='black', top='off')
plt.ylim([0,2.2])
plt.xlim([0,p_array[-1]*1.1])
plt.xticks(p_array,fontsize=14)
plt.yticks(fontsize=14)
plt.legend(loc='upper left', fontsize = 15)
plt.show()
```
